```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset = pd.read_csv('airline-price-prediction.csv')
dataset['date_dep_time'] = dataset.date.astype(str) + " " + dataset.dep_time
dataset.date_dep_time = pd.to_datetime(dataset.date_dep_time)
dataset['date_arr_time'] = dataset.date.astype(str) + " " + dataset.arr_time
dataset.date_arr_time = pd.to_datetime(dataset.date_arr_time)
dataset
```

```python
dataset.drop('date', axis=1,inplace= True)
dataset.drop('dep_time', axis=1,inplace= True)
dataset.drop('arr_time', axis=1,inplace= True)
dataset.drop('time_taken', axis=1,inplace= True)
dataset.drop('ch_code', axis=1,inplace= True)
dataset['price'] = pd.to_numeric(dataset['price'].str.replace(',',''), downcast="integer")
x = dataset['route'].str.replace("[\\W]+", ' ')
x = x.str.split(" ", expand=True)
dataset["src"] = x[2]
dataset['dest'] = x[4]
dataset.drop('route',axis = 1,inplace = True)
dataset
```

```python
dataset = dataset.iloc[:,:].values
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [3])],remainder= 'passthrough')
dataset = ct.fit_transform(dataset)
dataset = pd.DataFrame(dataset)
dataset
```

```python
dataset = dataset.iloc[:,:].values
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [2])],remainder= 'passthrough')
dataset = ct.fit_transform(dataset)
dataset = pd.DataFrame(dataset)
dataset
```

```python
dataset[11] = dataset[11].str.split('p', expand=True)[0].str.replace('+','',regex=True)
dataset = dataset.iloc[:,:].values
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [11])],remainder= 'passthrough')
dataset = ct.fit_transform(dataset)
dataset = pd.DataFrame(dataset)
dataset
```
